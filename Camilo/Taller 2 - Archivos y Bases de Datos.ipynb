Cree la base de datos (copie el código SQL que se usó)

```python
CREATE TABLE enfermedad
(
id_enfermedad int AUTO_INCREMENT PRIMARY KEY,
nombre varchar(255),
mapped_trait varchar (255),
URL varchar (255)
);

CREATE TABLE journal
(
id_jorunal int AUTO_INCREMENT PRIMARY KEY,
nombre varchar (255)
);

CREATE TABLE estudio
(
id_estudio int AUTO_INCREMENT PRIMARY KEY,
nombre varchar(255),
tamaño_rep int,
replicas int,
p_value int,
p_value_Text varchar (255),
or_bet int,
accesion varchar (255),
foreign key (id_jorunal) REFERENCES journal(id_jorunal)
foreign key (id_enfermedad) REFERENCES enfermedad(id_enfermedad)
);


CREATE TABLE publicacion(
id_pubmed int AUTO_INCREMENT PRIMARY KEY, 
fecha_pub datetime,
id_journal int,
link varchar (255),
CI varchar (255),
foreign key (id_estudio) REFERENCES estudio(id_estudio)
);
 

CREATE TABLE loci(
id_loci int AUTO_INCREMENT PRIMARY KEY,
region varchar(255),
chrom varchar(255), 
pos int,
genes_reportados int,
gen_mapped varchar(255),
gen_upstream int,
gen_downstream int,
snp_gen_id int,
ups_gen_dis int,
downs_gen_dis int,
snp_risk varchar (255),
snps varchar (255),
id_snp int,
contexto varchar (255),
intergenic binary,
allel_rskf int
);

CREATE TABLE enfermedad_loci
(
id_enfermedad int,
id_loci int,
PRIMARY KEY (id_enfermedad, id_loci),
foreign key (id_enfermedad) references enfermedad(id_enfermedad),
foreign key (id_loci) references loci(id_loci)
);

CREATE TABLE plataforma
(
id_plataforma int AUTO_INCREMENT PRIMARY KEY,
plataforma varchar(255)

);
```
